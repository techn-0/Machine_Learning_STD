```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets as ds

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

(x_train, y_train), (x_test, y_test) = ds.mnist.load_data()
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype(np.float32) / 255.0
x_test = x_test.astype(np.float32) / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

mlp = Sequential()
mlp.add(Dense(units=512, activation='sigmoid', input_shape=(784,))) # 은닉층 활성함수 시그노이드로 변경
mlp.add(Dense(units=10, activation='softmax'))

mlp.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.1), metrics=['accuracy']) # learning_rate 값 수정: 높은 발산에서의 결과를 보기위해 수정
mlp.fit(x_train, y_train, batch_size=128, epochs=50, validation_data=(x_test, y_test), verbose = 2)

res = mlp.evaluate(x_test, y_test, verbose=0)
print('정확률=', res[1] * 100)

```

```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets as ds

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD, Adam

(x_train, y_train), (x_test, y_test) = ds.mnist.load_data()
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype(np.float32) / 255.0
x_test = x_test.astype(np.float32) / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

#SGD
mlp_sgd = Sequential()
mlp_sgd.add(Dense(units=512, activation='sigmoid', input_shape=(784,))) # 은닉층 활성함수 시그노이드로 변경
mlp_sgd.add(Dense(units=10, activation='softmax'))

mlp_sgd.compile(loss='categorical_crossentropy', optimizer=SGD(learning_rate=0.1), metrics=['accuracy']) # learning_rate 값 수정: 높은 발산에서의 결과를 보기위해 수정
hist_sgd = mlp_sgd.fit(x_train, y_train, batch_size=32, epochs=50, validation_data=(x_test, y_test), verbose = 2) # 배치사이즈를 줄여 더 자주 업데이트 하게 함

res = mlp_sgd.evaluate(x_test, y_test, verbose=2)
print('SGD 정확률=', mlp_sgd.evaluate(x_test, y_test, verbose = 0) [1] * 100)
# Adam
mlp_adam = Sequential()
mlp_adam.add(Dense(units=512, activation='sigmoid', input_shape=(784,))) # 은닉층 활성함수 시그노이드로 변경
mlp_adam.add(Dense(units=10, activation='softmax'))

mlp_adam.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.1), metrics=['accuracy']) # learning_rate 값 수정: 높은 발산에서의 결과를 보기위해 수정
hist_adam = mlp_adam.fit(x_train, y_train, batch_size=32, epochs=50, validation_data=(x_test, y_test), verbose = 2) # 배치사이즈를 줄여 더 자주 업데이트 하게 함

res = mlp_adam.evaluate(x_test, y_test, verbose=2)
print('Adam 정확률=', mlp_adam.evaluate(x_test, y_test, verbose = 0) [1] * 100)

# 정확률 그래프
import matplotlib.pyplot as plt

plt.plot(hist_sgd.history['accuracy'], 'r--')
plt.plot(hist_sgd.history['val_accuracy'],'r')
plt.plot(hist_adam.history['accuracy'], 'b--')
plt.plot(hist_adam.history['val_accuracy'],'b')

plt.title('Comparison of SGD and Adam optimizers')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim((0.6, 1.0)) # 축범위 조절
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train_sgd', 'val_sgd', 'train_adam', 'val_adam'])
plt.grid()
plt.show()

# Loss그래프

plt.figure(figsize=(8,5)) # 그래프 가로, 세로 크기 조절

# 훈련 손실 (SGD)
plt.plot(hist_sgd.history['loss'], linestyle='--', label='train_sgd_loss')
# 검증 손실 (SGD)
plt.plot(hist_sgd.history['val_loss'],       label='val_sgd_loss')

# 훈련 손실 (Adam)
plt.plot(hist_adam.history['loss'], linestyle='--', label='train_adam_loss')
# 검증 손실 (Adam)
plt.plot(hist_adam.history['val_loss'],       label='val_adam_loss')

plt.title('Optimizer별 손실(Loss) 곡선 비교')      # 그래프 제목
plt.xlabel('Epochs')                            # x축 레이블
plt.ylabel('Loss')                              # y축 레이블
plt.legend()                                    # 범례 표시
plt.grid()
plt.show()

```

```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets as ds

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

# 다른 옵티마이저 임포트
from tensorflow.keras.optimizers import RMSprop, Adagrad

# 로드&전처리
(x_train, y_train), (x_test, y_test) = ds.mnist.load_data()
x_train = x_train.reshape(60000, 784).astype(np.float32) / 255.0
x_test  = x_test.reshape(10000, 784).astype(np.float32) / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test  = tf.keras.utils.to_categorical(y_test, 10)

#RMSprop
mlp_rms = Sequential()
mlp_rms.add(Dense(units=512, activation='sigmoid', input_shape=(784,)))
# 은닉층 활성함수 시그노이드로 변경
mlp_rms.add(Dense(units=10, activation='softmax'))

# optimizer를 RMSprop으로 변경
# learning_rate 값 수정: 높은 발산에서의 결과를 보기위해 수정
mlp_rms.compile(
    loss='categorical_crossentropy',
    optimizer=RMSprop(learning_rate=0.1),
    metrics=['accuracy']
)

hist_rms = mlp_rms.fit(
    x_train, y_train,
    batch_size=32, # 배치사이즈를 줄여 더 자주 업데이트 하게 함
    epochs=50,
    validation_data=(x_test, y_test),
    verbose=2
)

print('RMSprop 정확률=', mlp_rms.evaluate(x_test, y_test, verbose = 0) [1] * 100)

# Adagrad
mlp_ada = Sequential()
mlp_ada.add(Dense(units=512, activation='sigmoid', input_shape=(784,)))
mlp_ada.add(Dense(units=10, activation='softmax'))

# optimizer를 Adagrad로 변경
# learning_rate 값 수정: 높은 발산에서의 결과를 보기위해 수정
mlp_ada.compile(
    loss='categorical_crossentropy',
    optimizer=Adagrad(learning_rate=0.1),
    metrics=['accuracy']
)

hist_ada = mlp_ada.fit(
    x_train, y_train,
    batch_size=32, # 배치사이즈를 줄여 더 자주 업데이트 하게 함
    epochs=50,
    validation_data=(x_test, y_test),
    verbose=2
)
print('RMSprop 정확률=', mlp_ada.evaluate(x_test, y_test, verbose = 0) [1] * 100)

#정확도 그래프
import matplotlib.pyplot as plt

plt.plot(hist_rms.history['accuracy'], 'g--')
plt.plot(hist_rms.history['val_accuracy'], 'g')
plt.plot(hist_ada.history['accuracy'], 'm--')
plt.plot(hist_ada.history['val_accuracy'], 'm')

plt.title('RMSprop vs Adagrad 정확도')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim(0.6, 1.0) # 축범위 조절
plt.legend()
plt.grid()
plt.show()

# Loss 그래프
plt.figure(figsize=(8,5))
plt.plot(hist_rms.history['loss'],linestyle='--', label='train_rms_loss')
plt.plot(hist_rms.history['val_loss'], label='val_rms_loss')
plt.plot(hist_ada.history['loss'], linestyle='--', label='train_ada_loss')
plt.plot(hist_ada.history['val_loss'], label='val_ada_loss')

plt.title('RMSprop vs Adagrad Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```
