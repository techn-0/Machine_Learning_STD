```python
import cv2 as cv
from google.colab.patches import cv2_imshow

img = cv.imread('데저트이글.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

canny1 = cv.Canny(gray, 30, 100)
canny2 = cv.Canny(gray, 70, 180)

cv2_imshow(gray)
cv2_imshow(canny1)
cv2_imshow(canny2)

# cv.imshow('gray', gray)
# cv.imshow('canny1', canny1)
# cv.imshow('canny2', canny2)

# cv.waitKey()
# cv.destroyAllWindows()
```
